```python
names= ['fuvest_2020','fuvest_2019','fuvest_2018','fuvest_2017','fuvest_2016','fuvest_2015','fuvest_2014','fuvest_2013','fuvest_2012']
final_columns = ['CORTE_2020','CORTE_2019','CORTE_2018','CORTE_2017','CORTE_2016', 'CORTE_2015', 'CORTE_2014', 'CORTE_2013','CORTE_2012']
dataframes = []

WHITE = (255, 255, 255)
running = True

dataset = load(names, dataframes)
dataset_selected = []
for df in dataset:
    df = df.rename(index = df['CODIGO_E_NOME_DA_CARREIRA'])
    d = select(['CORTE_MIN'], df)
    d['CORTE_MIN'] = d['CORTE_MIN'].astype('float64')
    d = d.dropna()
    dataset_selected.append(d)
# print(dataset_selected)
```

```python
result = pd.concat(dataset_selected, axis=1, sort=False)
result.columns = final_columns


#result_without_new_courses = result.dropna(thresh=result.shape[1]-2, axis=0)
nomes = result.index

codes = nomes.str.extract('(\d+)')[0]
names = nomes.str.extract('\d+−(.*)')[0]
```

```python

cut_variances = result.var(axis = 1)

cut_sd = cut_variances ** (1/2)

cut_mean = result.mean(axis = 1)
```

```python
summary_dataframe = pd.DataFrame({"Código do curso":codes.values,"Nome do curso":names.values, "Média": cut_mean, "Desvio Padrão":cut_sd, 'Variância': cut_variances})
```

```python
summary_dataframe = summary_dataframe.sort_values('Variância')
summary_dataframe = summary_dataframe.dropna()
```
